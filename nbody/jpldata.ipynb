```python
from astroquery.jplhorizons import Horizons
from astropy.time import Time
import astropy.units as u
import re
object = Horizons(id='499', location='0', epochs={'start':'2023-11-03', 'stop':'2023-11-05',
                       'step':'3d'}).vectors(get_raw_response=True)

object_tab = Horizons(id='0', location='0', epochs={'start':'2023-11-03', 'stop':'2023-11-05',
                       'step':'3d'}).vectors()
m = re.search(r"\s+?Mass\s+?([\S]+).*?=\s+?([\S]+?)\s+?", object)
r = re.search(r'\s+?radius\s+?([\S]+).*?=\s+?([\S]+?)\s+?', object)
n = re.search(r"Target body name: (.+?) \((\d+)\)", object)



if m:  
    print("m=", m)
    mbase, mexp_s = m.groups()[1], m.groups()[0]
    print("r=", r)
    rad =r.groups()[1]
    name = n.groups()[0] + n.groups()[1]
else:
    print("Mass not found")

mass = float(mbase.split('+-')[0]) * 10**(float(mexp_s.split('x10^')[1]))
radius = float(rad.split('+-')[0])*1000
x, y, z = object_tab['x'].quantity[0].to_value(u.m), object_tab['y'].quantity[0].to_value(u.m), object_tab['z'].quantity[0].to_value(u.m)
vx, vy, vz = object_tab['vx'].quantity[0].to_value(u.m/u.s), object_tab['vy'].quantity[0].to_value(u.m/u.s), object_tab['vz'].quantity[0].to_value(u.m/u.s)
print(f'mass={mass}, rad={radius}, x={x}, y={y}, z={z}, vx={vx}, vy={vy}, vz={vz}')
print(name)

```
